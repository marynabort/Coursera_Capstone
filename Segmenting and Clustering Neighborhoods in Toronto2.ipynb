```python
#Read the table on the website
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
#save it as a new dataframe
my_df=tables[0]
my_df
```

```python
#remove all the rows that contain "Not assigned"
df = my_df[my_df.Borough != 'Not assigned']
df
```

```python
#replace "/" with "," in the last column
df['Neighborhood'] = df['Neighborhood'].str.replace(' /',',')
df
```

```python
#reset index
df = df.reset_index(drop=True)
df
```

```python
#check how many rows and columns are in the dataframe
df.shape
```

```python
#read the csv file with the coordinates into a dataframe
file_url='https://cocl.us/Geospatial_data'
lon_lat_df=pd.read_csv(file_url)
lon_lat_df
```

```python
#merge the two dataframes based on the postal code
dfinal = df.merge(lon_lat_df, how='inner', left_on='Postal code', right_on='Postal Code')
dfinal
```

```python
#drop the extra column and rename the 1st column like in the example
dfinal=dfinal.drop('Postal Code', axis=1)
dfinal = dfinal.rename(columns={'Postal code': 'PostalCode'})
dfinal
```
